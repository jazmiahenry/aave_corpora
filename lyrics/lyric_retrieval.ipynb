```python
#use chunk for multiple artists

artists = ['Mary J. Blige', 'Luther Vandross', 'Otis Redding', 'Ray Charles', 'James Brown', 'Al Green', 'Etta James', 'Chaka Khan'
, 'Aaliyah', 'Diana Ross', 'Prince', 'Rihanna', 'Tina Turner', 'Curtis Mayfield', 'Anita Baker', 'Patti LaBelle', 'Lionel Richie'
, 'Sade Adu', 'Babyface', 'Jill Scott', 'John Legend', 'Barry White', 'Donny Hathaway', 'Teddy Pendergrass', 'Maxwell', 'Dionne Warwick'
, 'D\'Angelo', 'Ne-Yo', 'T-Pain', 'Donna Summer', 'Keith Sweat', 'Tevin Campbell', 'Beyonce']

def get_lyrics(arr, k):  # Write lyrics of k songs by each artist in arr
    c = 0  # Counter
    for name in arr:
        try:
            songs = (genius.search_artist(name, max_songs = k, sort='popularity'))
            songs.save_lyrics()
        except: #  Broad catch which will give us the name of artist and song that threw the exception
            c += 1
            print(f"some exception at {name}: {c}")


get_lyrics(artists, 1000)
```

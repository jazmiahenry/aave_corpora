```python
#function to preprocess the books - reads in the files, lowercases, removes punctuation, removes weird symbols, and then prints it out
def PreProcess(x):
    lines = []
    for book in books:
        with open(book) as f:
            data = f.read()
            data = data.lower()          
            data = re.sub(r'\d+', '', data)
            newdata = re.sub(r'[^\w\s]','',data)
            data = re.sub('â,â,â', ' ', newdata, re.IGNORECASE)
            #print('\n',data)
            add = 'preprocessed'
            filename = '%s_%s.txt' % (book[13:-4], add)
        with open(filename, 'w') as f:
            f.writelines(data)
    print('Done!')
#the first two lines are the author and the book title            
```
